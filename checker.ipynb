```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
from mnga.rl.agents import DQNAgent, DDQNAgent, DDQNPERAgent, DDQNPERDuelingAgent, CategoricalAgent, RainbowAgent

env_id = "CartPole-v1"
env = gym.make(env_id)
```

```python
# DQNAgent
seed = 777 
np.random.seed(seed)
num_frames = 60000
memory_size = 10000
batch_size = 128
target_update = 3000
epsilon_decay = 0.9998

agent = DQNAgent(env, memory_size, batch_size, target_update, epsilon_decay, seed)
agent.train(num_frames)
```

```python
# DDQNAgent
seed = 777 
np.random.seed(seed)
num_frames = 60000
memory_size = 20000
batch_size = 128
target_update = 2000
epsilon_decay = 0.9998


agent_d = DDQNAgent(env, memory_size, batch_size, target_update, epsilon_decay, seed)
agent_d.train(num_frames)
```

```python
# DDQNPERAgent
seed = 777 
np.random.seed(seed)
num_frames = 60000
memory_size = 20000
batch_size = 128
target_update = 2000
epsilon_decay = 0.9998


agent_dper = DDQNPERAgent(env, memory_size, batch_size, target_update, epsilon_decay, seed)
agent_dper.train(num_frames)
```

```python
# DDQNPERDuelingAgent
seed = 777 
np.random.seed(seed)
num_frames = 60000
memory_size = 20000
batch_size = 128
target_update = 2000
epsilon_decay = 0.9998


agent_dueling = DDQNPERDuelingAgent(env, memory_size, batch_size, target_update, epsilon_decay, seed)
agent_dueling.train(num_frames)
```

```python
# CategoricalAgent
seed = 777 
np.random.seed(seed)
num_frames = 60000
memory_size = 50000
batch_size = 128
target_update = 1000
epsilon_decay = 0.99995


agent_categorical = CategoricalAgent(env, memory_size, batch_size, target_update, epsilon_decay, seed)
agent_categorical.train(num_frames)
```

```python
# RainbowAgent
seed = 778
np.random.seed(seed)

num_frames = 10_000
memory_size = 10_000
batch_size = 128
target_update = 100
n_step = 3


agent_rainbow = RainbowAgent(
    env,
    memory_size,
    batch_size,
    target_update,
    seed,
    n_step=n_step
)
agent_rainbow.train(num_frames)
```
